```python
from PIL import Image
from collections import Counter
import numpy as np
```

```python
def get_hist_image(path):
    im1 = Image.open(path)
    array1 = np.array(im1)
    flat_array_1 = array1.flatten()
    RH1 = Counter(flat_array_1)
    
    H1 = []
    for i in range(256):
        if i in RH1.keys():
            H1.append(RH1[i])
        else:
            H1.append(0)
    
    return H1
```

```python
def L2Norm(H1,H2):
    distance =0
    for i in range(len(H1)):
        distance += np.square(H1[i]-H2[i])
    return np.sqrt(distance)
```
